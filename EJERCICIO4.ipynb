# Funciones e Activacion
Las funciones de activacion cumplen un papel principal en el momento de ajustar la respuesta de una neurona ante un estimulo.
A continuacion, se presentan las funciones de activacion de mayor uso.

```python
import numpy as np
import matplotlib.pyplot as plt

# Funcion para mostrar el comportamiento de la funcion de activacion
def plot_activation_function(function):
    n = np.linspace(-5, 5, 100)
    y = [function(i) for i in n]

    plt.axhline(color = 'grey')
    plt.axvline(color = 'grey')
    plt.plot(n, y)
```

## Funcion de activacion lineal

$ y = f(n) = n $

esta funcion tiene un comportamiento lineal entre el estimulo de la neurona y su salida

```python
def linear(n):
    return n

plot_activation_function(linear)
```

## Funcion de activacion relu
$$ y = f(n) = \begin {cases} n, &\text {Si $n \geq 0$}\\ 0, &\text {Si $n<0$}
\end{cases} $$

Esta funcion de activacion tiene un comportamiento lineal cuando el estimulo es mayor o igual a 0. Si el estimulo es menor a 0, entonces la salida es igual a 0.

```python
def relu(n):
    return n if n >= 0 else 0

plot_activation_function(relu)
```

## Función de activación de escalón

$$ y = f(n) \begin {cases} 1, &\text {Si $n \geq 0$}\\ 0, &\text {Si $n<0$} \end{cases} $$

Esta función de activacion es una de las funciones más simples. Su propósito es modificar el estimulo a una salida binaria, dependiendo de si el estímulo supera un umbral específico.

```python
def step(n):
    return 1 if n >= 0 else 0

plot_activation_function(step)
```

## Funcion de activacion sigmoide

$$ y = f(n) = \frac{1}{1 + e^{-n}} $$

Esta funcion de activacion modifica los valores de n y los ajusta en el intervalo [0,1]. Es asi como, para cualquier valor del estimulo (por muy grande que sea), la funcion sigmoide no supera el valor de 1, y para valore de n menores a 0, no sera inferior a 0.

```python
def sigmoid(n):
    return 1 / (1 + np.e ** -n)

plot_activation_function(sigmoid)
```

## Funcion de activacion tangente hiperbolica 

$$ y = f(n) \frac{2}{1 + e^{-2n}} - 1 $$

Esta funcion de activacion, sus valores posibles de salida estan en el intervalo [-1,1].
Por esta razon, para cualquier valor de n la funcion tangente hiperebolica (tanh) no supera el valor de 1, mientras que, para valores de n numeros a 0, no sera menor ue -1.

```python
def tanh(n):
    return 2 / (1 + np.e ** (-2 *n)) - 1

plot_activation_function(tanh)
```

```python
import ipywidgets as widgets

# Diccionario con las funciones y sus descripciones
functions = {
    'lineal': {'function': linear},
    'Relu': {'function': relu},
    'Escalon': {'function': step},
    'Sigmoide': {'function': sigmoid},
    'Tangente hiperbolica': {'function': tanh}
}

# Funcion para mostrar la comparacion del comportamiento de las funciones de activacion.
def compare_functions(functions_to_compare):
    n = np.linspace(-5, 5, 100)

    colors = ['blue', 'red', 'green', 'orange', 'purple']

    for i, label in enumerate(functions_to_compare):
        y = [functions[label]['function'](i) for i in n]
        plt.plot(n, y, color = colors[i], label = label)

    plt.legend()
    plt.show()

selection = widgets.SelectMultiple(
    options = functions.keys(),
    value = ['Escalon', 'Sigmoide'],
    rows = 5,
    description = 'Comparar:'
)

widgets.VBox([widgets.interactive(compare_functions, functions_to_compare = selection)])
```
