# Actividad 2

Lee el dataset y crea una red neuronal multicapa (utilizando tensorflow) para predecir la valoración de cereal, tomando como entradas las calorías, las proteínas, las grasas, el sodio, la fibra, los carbohidratos, los azúcares, el potasio, las vitaminas, la caducidad, el peso y la tazas.

# Leer los datos

```python
import pandas as pd 
data = pd.read_csv('Datasets/cereal.csv')
data 
```

```python
X = data.iloc[:, :-1]
X
```

```python
# Dividir los datos en características (X) y etiquetas (y)
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Normalizar los datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crear la red neuronal
from tensorflow import keras
from keras.layers import Dense

model = keras.Sequential([
    Dense(128, activation='relu', input_shape=[X.shape[1]]),
    Dense(64, activation='relu'),
    Dense(1)  # Salida continua para regresión
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Entrenar el modelo
model.fit(X_scaled, y, epochs=100, batch_size=32)
```
